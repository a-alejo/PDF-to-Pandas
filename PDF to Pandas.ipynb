# PDF to Pandas
Based of - https://www.youtube.com/watch?v=DoyRTMlVFTU&list=WL&index=25

In the video, the person uses ChatGPT to write a code that can take a PDF and be able to read in inside of Python. Then inside of Python that data can be manipulated using Pandas

Misc: Check this out - https://github.com/danshorstein/pythonic-accountant

```python
import io
import requests
from PyPDF2 import PdfFileReader

def extract_text_from_pdf(url, page_number):
    # download the PDF file
    response = requests.get(url)

    # convert the bytes object to a file-like object
    pdf_file = io.BytesIO(response.content)

    # open the POF file using PyPDF2
    pdf_reader = PdfFileReader(pdf_file)

    # extract the text from the specified page
    page = pdf_reader.getPage(page_number)
    text = page.extractText()

    # return the extracted text
    return text

# specify the URL of the PDF file you want to download
url = 'https://www.tabs3.com/support/sample/apreports.pdf'

# specify the page number you want to extract text from // 
page_number = 16

# extract the text from the specified page
text = extract_text_from_pdf(url, page_number)

# print the extracted text
print(text)
```
